```python
#Getting path to direct the data as states
path = "pulse/data/aggregated/user/country/india/state/"
Agg_state_list = os.listdir(path)
Agg_state_list
```

```python
#Creating a Table Agrregated_user1
#Creating the dataframe
agg_user_col = {'State':[],'Year':[],'Quater':[],'Register_users':[],'App_opens':[]}
for i in Agg_state_list:
  p_i = path + i + "/"
  Agg_yr = os.listdir(p_i)
  for j in Agg_yr:
    p_j = p_i + j + "/"
    Agg_yr_list = os.listdir(p_j)
    for k in Agg_yr_list:
      p_k = p_j + k
      Data = open(p_k,'r')
      D = json.load(Data)
      Reg_users = D['data']['aggregated']['registeredUsers']
      App_opens = D['data']['aggregated']['appOpens']
      agg_user_col['Register_users'].append(Reg_users)
      agg_user_col['App_opens'].append(App_opens)
      agg_user_col['State'].append(i)
      agg_user_col['Year'].append(j)
      agg_user_col['Quater'].append(int(k.strip('.json')))
Aggregated_user1 = pd.DataFrame(agg_user_col)
Aggregated_user1
```

```python
#Now creating Aggregate_User2 
agg_user_col1 = {'State':[],'Year':[],'Quater':[],'Users_Brand':[],'Users_Count':[],'Users_percentage':[]}
for i in Agg_state_list:
  p_i = path + i + "/"
  Agg_yr = os.listdir(p_i)
  for j in Agg_yr:
    p_j = p_i + j + "/"
    Agg_yr_list = os.listdir(p_j)
    for k in Agg_yr_list:
      p_k = p_j + k
      Data = open(p_k,'r')
      D = json.load(Data)
      try:
        for agg_user in D['data']['usersByDevice']:
          Users_Brand = agg_user['brand']
          Users_Count = agg_user['count']
          Users_percentage = agg_user['percentage']
          agg_user_col1['Users_Brand'].append(Users_Brand)
          agg_user_col1['Users_Count'].append(Users_Count)
          agg_user_col1['Users_percentage'].append(Users_percentage)
          agg_user_col1['State'].append(i)
          agg_user_col1['Year'].append(j)
          agg_user_col1['Quater'].append(int(k.strip('.json')))
      except:
        pass  
Aggregated_user2 = pd.DataFrame(agg_user_col1)
Aggregated_user2
```

```python
#Merging above two tables into single table as Aggregated_user
Aggregated_user = pd.merge(Aggregated_user1,Aggregated_user2,on=['State','Year','Quater'])
Aggregated_user
```

```python
#Creating second table Aggregated_transaction
path1 = "pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list1 = os.listdir(path1)

agg_trans_col = {'State':[],'Year':[],'Quater':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}
for i in Agg_state_list1:
  p_i = path1 + i + "/"
  Agg_yr = os.listdir(p_i)
  for j in Agg_yr:
    p_j = p_i + j + "/"
    Agg_yr_list = os.listdir(p_j)
    for k in Agg_yr_list:
      p_k = p_j + k
      Data = open(p_k,'r')
      D = json.load(Data)
      for z in D['data']['transactionData']:
        Name = z['name']
        count = z['paymentInstruments'][0]['count']
        amount = z['paymentInstruments'][0]['amount']
        agg_trans_col['Transaction_type'].append(Name)
        agg_trans_col['Transaction_count'].append(count)
        agg_trans_col['Transaction_amount'].append(amount)
        agg_trans_col['State'].append(i)
        agg_trans_col['Year'].append(j)
        agg_trans_col['Quater'].append(int(k.strip('.json')))
Aggregated_transaction = pd.DataFrame(agg_trans_col)
Aggregated_transaction
```

```python
#Creating 4th table Map_user
path3 = "pulse/data/map/user/hover/country/india/state/"
Agg_state_list3 = os.listdir(path3)

map_user_col = {'State':[],'Year':[],'Quater':[],'District':[],'Registered_Users':[],'App_Opens':[]}
for i in Agg_state_list3:
  p_i = path3 + i + "/"
  Agg_yr = os.listdir(p_i)
  for j in Agg_yr:
    p_j = p_i + j + "/"
    Agg_yr_list = os.listdir(p_j)
    for k in Agg_yr_list:
      p_k = p_j + k
      Data = open(p_k,'r')
      D = json.load(Data)
      user_map = D['data']['hoverData']
      for district,values in user_map.items():
        District = district
        Registered_Users = values['registeredUsers']
        App_Opens = values['appOpens']
        map_user_col['District'].append(District)
        map_user_col['Registered_Users'].append(Registered_Users)
        map_user_col['App_Opens'].append(App_Opens)
        map_user_col['State'].append(i)
        map_user_col['Year'].append(j)
        map_user_col['Quater'].append(int(k.strip('.json')))
Map_user = pd.DataFrame(map_user_col)
Map_user
```

```python
#Connect Python to DataBase
mydb = pymysql.connect(
    host = "localhost",
    user = "root",
    password="root",
    database = "phonepe_trans_insights",
    autocommit=True
)

cursor = mydb.cursor()
print("Database Connected")
```

```python
#SQL code to create 1st table Aggregated_user

cursor.execute("DROP TABLE IF EXISTS Aggregated_user")

table1 = """
        CREATE TABLE IF NOT EXISTS Aggregated_user(
            State varchar(100),
            Year INT,
            Quater INT,
            Register_users BIGINT,
            App_opens BIGINT,
            Users_Brand varchar(100),
            Users_Count INT,
            Users_percentage FLOAT
        );
"""

cursor.execute(table1)
print("Table Created")    
```

```python
#Inserting the values into the table aggregated_user

for i,row in Aggregated_user.iterrows():
    sql1 = """
            INSERT INTO Aggregated_user(
                State,Year,Quater,Register_users,App_opens,Users_Brand,Users_Count,Users_percentage
            )
            VALUES (%s,%s,%s,%s,%s,%s,%s,%s)
        """
    values1 = (row['State'],row['Year'],row['Quater'],row['Register_users'],row['App_opens'],row['Users_Brand'],row['Users_Count'],row['Users_percentage'])
    cursor.execute(sql1,values1)

print("Data inserted Sucessfully")
```
